### $\bullet$ Web scraping

[Back to top](#plan)

```python
# scraping using requests
def requestsHTML(url):
    try:
        html = requests.get(url, timeout = 5).text
    except:
        html = ''
    return html


# scraping using selenium
def seleniumHTML(url, path_to_chrome_driver):
    # open browser
    browser = webdriver.Chrome(path_to_chrome_driver)
    
    # collect content
    try:
        browser.get(url)
        html  = browser.page_source
        title = browser.title
    except:
        html  = ''
        title = ''
    
    # close browser
    browser.quit()
    return (html, title)


# scraping using trafilatura
def trafilaturaHTML(url):
    html = trafilatura.fetch_url(url)
    if not html: html = ''
    return html
```

```python
url = 'https://en.wikipedia.org/wiki/Data_science'
```

```python
html_selenium, title_selenium = seleniumHTML(url, path_to_chrome_driver)
```

### $\bullet$ URL scraping

[Back to top](#plan)

```python
hrefs = soup_selenium.find_all(href = True)

hrefs = [h['href'] for h in hrefs]
hrefs
```

```python
url_wiki = 'https://en.wikipedia.org'
```

```python
imgs = soup_selenium.find_all(attrs = {'class': 'image'}, href = True)
for img in imgs : print(url_wiki + img['href'])
```

```python
# find all subsubheader
h3 = soup_selenium.find_all('h3')

# find 'Techniques' subsubheader 
h3_techniques = [h.find_all('span', attrs = {'id': 'Techniques'}) for h in h3]
h3_techniques = [h for hs in h3_techniques for h in hs][0]

#print(h3_techniques)

# get list of 'a' in next container
techniques = h3_techniques.find_next('ul').find_all('a', href = True, title = True)

#print(techniques)

# get list of hyperlinks
techniques = [url_wiki + t['href'] for t in techniques]
for t in techniques:
    print(t)
```
